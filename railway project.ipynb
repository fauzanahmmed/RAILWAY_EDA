# peak hours and days for ticket purchases

```python
# Converting the Timestamp Column to a Datetime Object
rt['Time_of_Purchase'] = pd.to_datetime(rt['Time_of_Purchase'])



# Extracting day of the week and hour from the timestamp
rt['day_of_week'] = rt['Time_of_Purchase'].dt.day_name()
rt['hour'] = rt['Time_of_Purchase'].dt.hour




# Analyzing Peak Days
# Count purchases by day
day_counts = rt['day_of_week'].value_counts()

# Print or visualize the results
print(day_counts)
day_counts.plot(kind='bar', title='Ticket Purchases by Day of the Week', ylabel='Number of Purchases', xlabel='Day of the Week')




# Analyzing Peak Hours
# Count purchases by hour
hour_counts = rt['hour'].value_counts().sort_index()

# visualizing the results
print(hour_counts)
hour_counts.plot(kind='line', title='Ticket Purchases by Hour', ylabel='Number of Purchases', xlabel='Hour of the Day')



# Combining Analysis
# Group by both day and hour
day_hour_counts = rt.groupby(['day_of_week', 'hour']).size().unstack()


# Heatmap of combined analysis
plt.figure(figsize=(16, 10))
sns.heatmap(day_hour_counts, cmap='Blues', annot=True, fmt='d')
plt.title('Ticket Purchases by Day and Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Day of the Week')
plt.show()
```

# seasonal trends in ticket purchases

```python
# Loading the dataset
file_path = 'D:/datascience/project/data/railway.csv'
rt = pd.read_csv(file_path)

# Converting 'Date of Purchase' to datetime format
rt['Date of Purchase'] = pd.to_datetime(rt['Date of Purchase'])

# Extracting the month and year for trend analysis
rt['Year'] = rt['Date of Purchase'].dt.year
rt['Month'] = rt['Date of Purchase'].dt.month_name()

# finding Aggregate ticket purchases by month (ignoring year for seasonal trends)
monthly_trends = rt.groupby('Month').size().reindex(
    ['January', 'February', 'March', 'April', 'May', 'June', 
     'July', 'August', 'September', 'October', 'November', 'December']
)

# Ploting the monthly trends
plt.figure(figsize=(12, 6))
sns.barplot(x=monthly_trends.index, y=monthly_trends.values, palette="Blues_d")
plt.title("Seasonal Trends in Ticket Purchases (Monthly Aggregation)", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Number of Tickets Purchased", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

# average ticket price across different ticket classes and types

```python
# Grouping by 'Ticket Class' and 'Ticket Type' and calculating the mean price
average_price = rt.groupby(['Ticket Class', 'Ticket Type'])['Price'].mean().reset_index()

# Renameing columns for clarity
average_price.columns = ['Ticket Class', 'Ticket Type', 'Average Price']

# Displaying the result
print(average_price)
```

# significant price difference between online and station purchases

```python


# Grouping the data by 'Purchase Type' and calculating the average price for each type
average_prices = rt.groupby('Purchase Type')['Price'].mean()

# Calculating the price difference between the two types
price_difference = abs(average_prices['Online'] - average_prices['Station'])

# Printing the results
print("Average Prices:")
print(average_prices)
print(f"\nPrice Difference: {price_difference:.2f}")
```

# routes (departure to destination) are most common

```python
# Ensuring the necessary columns are present
if {'departure', 'destination'}.issubset(rt.columns):
    # Counting the occurrences of each route
    route_counts = rt.groupby(['departure', 'destination']).size().reset_index(name='count')

    # Sorting by count in descending order
    most_common_routes = route_counts.sort_values(by='count', ascending=False)

    # Displaying the top 10 most common routes
    print("Top 10 most common routes:")
    print(most_common_routes.head(10))
else:
    print("The dataset does not contain the required 'departure' and 'destination' columns.")
```

# percentage of journeys experience delays

```python
# Calculating the total number of journeys
total_journeys = rt.shape[0]

# Calculating the number of delayed journeys
delayed_journeys = rt[rt['Journey Status'] == 'Delayed'].shape[0]

# Calculating the percentage of delayed journeys
delayed_percentage = (delayed_journeys / total_journeys) * 100

total_journeys, delayed_journeys, delayed_percentage

# specifing output
print(f"Total Journeys: {total_journeys}")
print(f"Delayed Journeys: {delayed_journeys}")
print(f"Percentage of Journeys Delayed: {delayed_percentage:.2f}%")

```

# are the most common reasons for delays

```python
# Assuming the CSV has a column named 'Delay Reason'
# Check if the column exists
if 'Delay Reason' in rt.columns:
    # Count the occurrences of each reason
    delay_reasons = rt['Delay Reason'].value_counts()

    # Display the most common reasons
    print("Most Common Reasons for Delays:")
    print(delay_reasons.head(10))
else:
    print("The column 'Delay Reason' does not exist in the dataset.")
```

# percentage of delayed journeys resulted in a refund request

```python
# Filtering for delayed journeys
delayed_journeys = rt[rt['Journey Status'] == 'Delayed']

# Calculating total delayed journeys and refund requests
total_delayed = len(delayed_journeys)
refund_requests = delayed_journeys[delayed_journeys['Refund Request'] == 'Yes']
total_refund_requests = len(refund_requests)

# Calculating the percentage of delayed journeys with refund requests
percentage_refund_requests = (total_refund_requests / total_delayed) * 100 if total_delayed > 0 else 0

# Printing the specified output
print(f"Total Delayed Journeys: {total_delayed}")
print(f"Total Refund Requests for Delayed Journeys: {total_refund_requests}")
print(f"Percentage of Delayed Journeys with Refund Requests: {percentage_refund_requests:.2f}%")

```

# patterns in refund requests based on journey status or delay reason

```python
# Grouping by Journey Status and calculating the percentage of refund requests
journey_status_analysis = (
    rt.groupby("Journey Status")["Refund Request"]
    .value_counts(normalize=True)
    .unstack()
    .fillna(0)
    * 100
)

# Grouping by Reason for Delay and calculating the percentage of refund requests
reason_delay_analysis = (
    rt[rt["Reason for Delay"].notnull()]
    .groupby("Reason for Delay")["Refund Request"]
    .value_counts(normalize=True)
    .unstack()
    .fillna(0)
    * 100
)

# Displaying the results
journey_status_analysis, reason_delay_analysis.head()
```

# does the use of railcards impact ticket prices

```python
# Group by 'Railcard' and calculate the average ticket price
average_prices = rt.groupby('Railcard')['Price'].mean().sort_values()

# Plot the results
plt.figure(figsize=(10, 6))
average_prices.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Ticket Price by Railcard Usage', fontsize=16)
plt.xlabel('Railcard Type', fontsize=14)
plt.ylabel('Average Ticket Price ()', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
